```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import (
    accuracy_score,
    f1_score,
    recall_score,
    classification_report,
    ConfusionMatrixDisplay
)
```

```python
DATA_PATH = Path("../data/train.csv")

df = pd.read_csv(DATA_PATH)
X = df["text"].astype(str)
y = df["label"]
print("Total examples:", len(X))
print("Classes:", sorted(y.unique()))
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=30
)

print("Training examples:", X_train.shape[0])
print("Validation examples:", X_val.shape[0])

print("\nTrain class distribution:")
print((y_train.value_counts(normalize=True) * 100).round(2))

print("\nValidation class distribution:")
print((y_val.value_counts(normalize=True) * 100).round(2))
```

```python
variant_3_model = Pipeline(
    steps=[
        (
            "tfidf",
            TfidfVectorizer(
                lowercase=True,
                stop_words="english",
                ngram_range=(1, 2)
            )
        ),
        (
            "classifier",
            LogisticRegression(
                max_iter=1000,
                class_weight="balanced",
                random_state=30
            )
        )
    ]
)

variant_3_model
```

```python
variant_3_model.fit(X_train, y_train)
```

```python
y_pred_v3 = variant_3_model.predict(X_val)

print("Predictions completed for Version 3.")
print("Number of validation predictions:", len(y_pred_v3))
```

```python
accuracy_v3 = accuracy_score(y_val, y_pred_v3)

macro_f1_v3 = f1_score(
    y_val,
    y_pred_v3,
    average="macro"
)

weighted_f1_v3 = f1_score(
    y_val,
    y_pred_v3,
    average="weighted"
)

fraud_recall_v3 = recall_score(
    y_val,
    y_pred_v3,
    labels=["fraud-report"],
    average="macro",
    zero_division=0
)

print("Version 3 Results")
print("-----------------")
print("Model: TF-IDF unigrams + bigrams + Logistic Regression")
print("Class weighting: balanced")
print("Accuracy:", round(accuracy_v3, 4))
print("Macro F1:", round(macro_f1_v3, 4))
print("Weighted F1:", round(weighted_f1_v3, 4))
print("Fraud-report recall:", round(fraud_recall_v3, 4))
```

```python
print("Classification Report - Version 3")
print(classification_report(y_val, y_pred_v3))
```

```python
labels = sorted(y.unique())

ConfusionMatrixDisplay.from_predictions(
    y_val,
    y_pred_v3,
    labels=labels,
    normalize="true",
    xticks_rotation=45
)

plt.title("Confusion Matrix - Version 3")
plt.tight_layout()
plt.show()
```
